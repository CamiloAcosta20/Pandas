```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Archivos para carga de datos y gráficos/sampledata.csv")  # Ruta del archivo

# Asegurarse de que las columnas relevantes sean numéricas
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(55).astype(int)

# Crear un nuevo rango de edades para agrupar (opcional)
bins = [0, 18, 35, 50, 65, 100]
labels = ['0-18', '19-35', '36-50', '51-65', '66+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Agrupar por país y grupo de edades
df_grouped = df.groupby(['country', 'age_group']).size().unstack(fill_value=0)

# Filtrar los 10 países con más personas
df_grouped_filtered = df_grouped.loc[df_grouped.sum(axis=1).nlargest(10).index]

# Crear el gráfico de columnas apiladas
df_grouped_filtered.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 8),
    color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
)

# Etiquetas y título
plt.title("Distribución de Edades por los 10 Países con Más Personas")
plt.xlabel("País")
plt.ylabel("Número de Personas")
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas y alinearlas a la derecha
plt.legend(title="Grupos de Edad")
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python


import pandas as pd

def change_column_types(file_path, sheet_name, column_types):
    """
    Changes the data types of specified columns in an Excel sheet and returns a DataFrame.

    Args:
        file_path (str): The path to the Excel file.
        sheet_name (str): The name of the sheet to modify.
        column_types (dict): A dictionary where keys are column names and values are the desired data types.

    Returns:
        pandas.DataFrame: A DataFrame with the updated column types, or None if an error occurs.
    """
    try:
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        for column, dtype in column_types.items():
            if column in df.columns:
                try:
                    df[column] = df[column].astype(dtype)
                except (ValueError, TypeError) as e:
                    print(f"Error converting column '{column}' to type '{dtype}': {e}")
                    # Handle the error as needed (e.g., skip the column or return None)
                    # For this example we'll continue to the next column
                    continue
            else:
                print(f"Warning: Column '{column}' not found in the DataFrame")

        return df  # Return the modified DataFrame
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None
```

```python
# crear un dataframe para cada hoja de excel

import pandas as pd

# Cargar el archivo de Excel
file_path = '/content/drive/MyDrive/Colab Notebooks/Archivos para carga de datos y gráficos/Tendencia_Tic_2022.xlsx'
try:
    excel_data = pd.ExcelFile(file_path)
except FileNotFoundError:
    print(f"Error: Archivo no encontrado en la ruta '{file_path}'. Verifica la ruta.")
    exit()

# Obtener las hojas del libro
sheet_names = excel_data.sheet_names

# Crear un diccionario para almacenar los DataFrames de cada hoja
dfs = {}

# Iterar sobre las hojas y crear un DataFrame para cada una
for sheet_name in sheet_names:
    try:
        dfs[sheet_name] = pd.read_excel(excel_data, sheet_name=sheet_name)
        print(f"DataFrame para la hoja '{sheet_name}' creado exitosamente.")
    except ValueError as e:
        print(f"Error al leer la hoja '{sheet_name}': {e}")

# Ahora puedes acceder a cada DataFrame usando su nombre de hoja como clave en el diccionario 'dfs'
# Ejemplo:
# df_c1 = dfs['C.1']
# df_c2 = dfs['C.2']
# ... y así sucesivamente
```

```python
# unificar los dataframe de las hojas excepto el de la hoja 'Contenido'

# Concatenar todos los DataFrames excepto el de la hoja 'Contenido'
dfs_combined = pd.concat([dfs[sheet_name] for sheet_name in sheet_names if sheet_name != 'Contenido'], ignore_index=True)

# Mostrar el DataFrame combinado
dfs_combined
```

```python
# filtrar por los departamentos de ANTIOQUIA y CHOCÓ

# Assuming 'df' is your DataFrame and it has a 'Departamento' column
# Filter the DataFrame for the specified departments
filtered_df = dfs_combined[dfs_combined['DEPARTAMENTO'].isin(['ANTIOQUIA', 'CHOCÓ'])]

# Now 'filtered_df' contains only the rows where the 'Departamento' is 'ANTIOQUIA' or 'CHOCÓ'

# Example usage: print the filtered DataFrame
filtered_df
```
